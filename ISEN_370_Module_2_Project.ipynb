#Confidence Interval

Generate the 95% confidence interval for the mean estimated flow rate of Septemebr 2017.


````python

stats.norm.interval(
    confidence=0.95,
    loc=np.mean(lewas_sep["Est. Flow Rate[m^3/s]"]),
    scale=stats.sem(lewas_sep["Est. Flow Rate[m^3/s]"])
)

````

**Interpret the confidence interval you generated.**<br>

We are 95% confident that the true mean estimated flow rate for September 2017 lies between â€“0.000589 mÂ³/s and 0.000604 mÂ³/s.

Because the interval includes 0, it indicates that the average flow rate is effectively zero, suggesting very low or nearly stagnant water flow throughout most of the month. The small mean value and narrow range show that flow conditions were relatively stable, with only minor fluctuations likely caused by short rainfall events or measurement noise.

In practical terms, this confidence interval tells us that the September 2017 streamflow was minimal, and any variation observed in the data likely reflects natural variability rather than sustained flow changes.

# Hypothesis testing
Suppose we are interested in testing if the mean estimated flow rate from September 2017 is statisticlly different from the mean estimated flow rate from October 2017.

***Question***<br>
**How many samples are there in this experiment? What are they**?<br>

There are two samples in this experiment. The first sample is the estimated flow rate data collected in September 2017, and the second sample is the estimated flow rate data collected in October 2017. Each sample represents one month of flow measurements, and the goal is to compare their means to see if the average flow rate in September is statistically different from the average flow rate in October.


# Independent Variable and Dependent Variable

***Question***<br>
**What is the independent variable and what is the dependent variable of this experiment?**




***Answer***

The independent variable in this experiment is time period, specifically the month in which the data was collected (September 2017 or October 2017). The dependent variable is the estimated flow rate (mÂ³/s), since it is the value being measured and compared between the two months.

In other words, the month determines the data group, and the flow rate is the outcome that depends on that factor.

***Question***<br>
What are the assumptions for the two sample t test?

***Answer***

The two-sample t-test assumes that the two samples are independent, meaning that the data from September and October were collected separately and do not influence each other. It also assumes that the data in each sample are approximately normally distributed. Lastly, it assumes equal variances between the two groups; however, if the variances are not equal, the Welchâ€™s t-test can be used instead since it does not require this assumption.

#Hypothesis testing
Use the appropraite statistical analysis technique.<br>
Conduct model adequacy and comment on the model adequacy.<br>
Conduct hypothesis testing

````python
sept_flow = lewas_sep["Est. Flow Rate[m^3/s]"].dropna()
oct_flow = lewas_oct["Estimated Flow Rate (m3/s)"].dropna()

shapiro_sept = stats.shapiro(sept_flow.sample(500, random_state=1))
shapiro_oct = stats.shapiro(oct_flow.sample(500, random_state=1))
print("Shapiro-Wilk Normality Test:")
print(f"September p-value = {shapiro_sept.pvalue:.5f}")
print(f"October p-value   = {shapiro_oct.pvalue:.5f}")
````

````python
levene_test = stats.levene(sept_flow, oct_flow)
print("\nLeveneâ€™s Test for Equal Variances:")
print(f"Statistic = {levene_test.statistic:.5f}, p-value = {levene_test.pvalue:.5f}")
````

````python
t_stat, p_val = stats.ttest_ind(sept_flow, oct_flow, equal_var=False)
print("\nWelchâ€™s Two-Sample t-Test:")
print(f"t-statistic = {t_stat:.5f}")
print(f"p-value = {p_val:.5f}")
````

Based on the tests, what can you conclude about the model assumptions?
<br>
**Answer:**

Based on the Shapiro-Wilk test results, both September and October have p-values of 0.00000, which means we reject the null hypothesis of normality. This indicates that the data are not normally distributed. However, because both datasets are large, the Central Limit Theorem allows the use of the t-test, so this violation does not significantly affect the validity of the analysis.

The Leveneâ€™s test also has a p-value of 0.00000, showing that the variances of the two samples are not equal. This means the assumption of equal variances is violated. For this reason, the Welchâ€™s t-test (with equal_var=False) is the correct test to use because it does not assume equal variances.

Overall, while the normality and equal variance assumptions are not met, the model is still adequate due to the large sample size and the robustness of the Welchâ€™s t-test.

# 6-Step Hypothesis Testing


**Step 1: State the hypotheses**

The null hypothesis (Hâ‚€) says that the mean estimated flow rate in September 2017 is the same as the mean estimated flow rate in October 2017.
The alternative hypothesis (Hâ‚) says that the mean estimated flow rate in September 2017 is different from the mean estimated flow rate in October 2017.

**Step 2: Choose the significance level**

For this test, I used a 5% significance level (Î± = 0.05).

**Step 3: Select the test and check assumptions**

The Shapiro-Wilk test showed that the data are not normally distributed, and Leveneâ€™s test showed that the variances are not equal. Because of this, the Welchâ€™s two-sample t-test was the best choice since it doesnâ€™t assume equal variances and works well with large samples.

**Step 4: Calculate the test statistic and p-value**

The test gave a t-statistic of -22.20304 and a p-value of 0.00000.

**Step 5: Make a decision**

Since the p-value is less than 0.05, I rejected the null hypothesis.

**Step 6: Conclusion**

There is enough evidence to say that the average estimated flow rate in September 2017 is significantly different from the average estimated flow rate in October 2017. This suggests that the flow conditions changed a lot between the two months, probably because of more rainfall or storm activity in October.
